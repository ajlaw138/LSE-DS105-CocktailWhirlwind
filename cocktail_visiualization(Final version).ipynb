```python
import nltk.tokenize as tk

type_com_data = type_data.dropna(subset=['Composition'],axis=0)

def split_count(t):
    text = ' '.join(list(t))
    # Split words by text punctuation divider
    punctTokenizer = tk.WordPunctTokenizer()
    word_list = punctTokenizer.tokenize(text)
    Word_text = list(filter(lambda x:x not in [',','.','!','-',';','?',"'",'&','),',"(",")","’",":"],
                            word_list))

    count_text = len(pd.value_counts(Word_text).tolist())
    return count_text

Type = type_com_data.groupby("Type")['Composition'].aggregate(split_count).index.tolist()
Num = type_com_data.groupby("Type")['Composition'].aggregate(split_count).to_list()
Count = type_com_data.groupby("Type").count()['Price']
num_type_data = pd.DataFrame({'Type':Type,'Num_Count':np.array(Num)/np.array(Count)})


(ggplot(num_type_data,aes(x='Type',y='Num_Count',fill='Type'))+
    geom_bar(stat='identity')+
    labs(x='Cocktail Type',y='Number of Composition/Cocktail')+
    theme_bw()+
    theme(axis_text_x=element_text(angle=60))
)
```

```python
#Composition
Comp_text = ' '.join(list(type_com_data.Composition))
# Split words by text punctuation divider
punctTokenizer = tk.WordPunctTokenizer()
word_list = punctTokenizer.tokenize(Comp_text)
Word_text = list(filter(lambda x:x not in [',','.','!','-',';','?',"'",'&','),',"(",")","’",":"],word_list))
wc_text = ' '.join(Word_text)

import wordcloud
from PIL import Image

img = Image.open('E:/cocktail1.jpg') #Open the picture
img_array = np.array(img) #Convert the image to an array

wc = wordcloud.WordCloud(
    background_color='white',
    width=2000,
    height=1500,
    scale=2,
    min_font_size=1,
    mask=img_array #Set background picture
    #font_path=font,
)
wc.generate(wc_text)

plt.imshow(wc, interpolation='bilinear')# Show the picture with plt
plt.axis("off")  # Do not show coordinate axes
plt.show() # Show the picture

wc.to_file("E:/cocktail_type.jpg")
```

```python
comp = pd.value_counts(Word_text).index[:6].tolist()

def compute_price(word):
    price_index = list(map(lambda index:index if word in type_com_data.Composition[index] else None,
                     range(len(type_com_data))))
    #Delete None
    Price_index = filter(None,price_index)
    # Compute the average price
    average_price = np.mean(type_com_data.Price[Price_index])
    return round(average_price,2)

comp_price = list(map(compute_price,comp))
```

```python
# Graph
comp_price_data = pd.DataFrame({'Comp':comp,"Price":comp_price})
(ggplot(comp_price_data,aes(x='Comp',y='Price',fill='Comp'))+
    geom_bar(stat='identity')+
    labs(x='Composition',y='The average price of Composition')+
    theme_bw()+
    theme(axis_text_x=element_text(angle=60))
)
```

# Bootstrap

```python
import random
```

```python
word_dict = dict()
for i in range(1000):
    index = random.sample(range(len(type_com_data)),round(len(type_com_data)*0.8))
    data = type_com_data.iloc[index,:]
    Comp_text = ' '.join(list(data.Composition))
    punctTokenizer = tk.WordPunctTokenizer()
    word_list = punctTokenizer.tokenize(Comp_text)
    wordtext = list(filter(lambda x:x not in [',','.','!','-',';','?',"'",'&','),',"(",")","’",":"],word_list))
    comp = pd.value_counts(wordtext).index.tolist()
    data.Composition.reset_index(drop=True,inplace=True)
    data.Price.reset_index(drop=True,inplace=True)
    for word in comp:
        price_index = list(map(lambda x:x if word in data.Composition[x] else None,
                     range(len(data))))
        #Delete None
        Price_index = filter(None,price_index)
        # Compute the average price
        average_price = np.mean(data.Price[Price_index])
        if not word_dict.get(word) and word_dict.get(word) != []:
            word_dict[word] = []
        else:
            word_dict[word].append(average_price)
```

```python
graph_word_dict = dict(list(word_dict.items())[:6])
Boot_com_data = pd.DataFrame(graph_word_dict).melt()

(ggplot(Boot_com_data,aes(x='variable',y='value',fill='variable'))+
    geom_violin()+
    labs(x='Composition',y="Price Distribution Based on Bootstrap Method")+
    theme_bw()
)
```
